# Melt() 

```python
import pandas as pd

var1 = pd.DataFrame({"Days":[1,2,3,4,5,6],"Eng":[12,34,55,66,45,67],"Maths":[13,25,34,56,78,89]})
var1
```

```python
pd.melt(var1)
```

```python
pd.melt(var1,id_vars = ["Eng"])
```

```python
import pandas as pd
pd.melt(var1,id_vars = ["Days"],var_name = "python",value_name = "nidhi")
```

# Pivot()

```python
import pandas as pd
var1 = pd.DataFrame({"Days":[1,2,3,4,5,6],"name":['A','B','C','A','B','C'],"Eng":[12,34,55,66,45,67],"Maths":[13,25,34,56,78,89]})
var1
```

```python
var1.pivot(index = "Days",columns = "name", values = "Eng")
```

```python
var1.pivot_table(index = "Days",columns = "name", aggfunc = "mean",margins= "True")
```
